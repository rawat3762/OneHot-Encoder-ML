```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit_transform(en_data)
```

```python
ohe.fit_transform(en_data).toarray()
```

```python
ar = ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns = ['Gender_Female','Gender_Male','Married_No','Married_Yes'])
```

```python
ohe = OneHotEncoder(drop = 'first')
```

```python
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
new_dataset = pd.DataFrame(ar,columns = ['Gender_Male','Married_Yes'])
```

```python
dataset[['Gender','Married']] = new_dataset
dataset
```
